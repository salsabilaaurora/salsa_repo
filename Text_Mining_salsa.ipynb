```python
import os
from docx import Document
from text_processing import *
import string

def preprocess_text(text):
    # Mengubah teks menjadi huruf kecil
    text = text.lower()

    # Menghapus karakter khusus dan tanda baca
    text = ''.join([char for char in text if char not in string.punctuation])

    # Menghapus spasi berlebih
    text = ' '.join(text.split())

    return text

from collections import Counter

# Fungsi untuk membagi teks menjadi token-token (kata-kata)
def tokenize_text(text):
    tokens = text.split()  # Anda dapat menggunakan pemisah kata yang sesuai
    return tokens

# Fungsi untuk menghitung frekuensi kemunculan kata-kata dalam teks
def calculate_word_frequency(tokens):
    word_frequency = Counter(tokens)
    return word_frequency

# Fungsi untuk membaca teks dari berkas DOCX
def read_text_from_docx(file_path):
    doc = Document(file_path)
    text = ""
    for paragraph in doc.paragraphs:
        text += paragraph.text + "\n"
    return text

# Fungsi untuk menyimpan hasil frekuensi kata-kata ke berkas teks (txt)
def save_word_frequency_to_txt(frequency_dict, file_path):
    with open(file_path, 'w') as file:
        file.write('Kata\tFrekuensi\n')  # Header

        # Isi file dengan data frekuensi kata-kata
        for word, freq in frequency_dict.items():
            file.write(f'{word}\t{freq}\n')

absen = input("Berapa Absenmu: ")

# Absen yang akan diinput
target_absen = f"33226000{absen}"

# Direktori utama (folder "Source")
base_directory = "C:/Users/user/Untitled Folder/Source"

# Daftar folder yang ingin diperiksa
folders_to_check = ['Kesehatan', 'Teknologi', 'Politik', 'Olahraga', 'Kriminal']

# Loop melalui setiap folder
for folder_name in folders_to_check:
    folder_path = os.path.join(base_directory, folder_name)

    # Loop melalui setiap file dalam folder
    for filename in os.listdir(folder_path):
        if target_absen in filename:
            # File dengan nama yang mirip ditemukan dalam folder ini
            file_path = os.path.join(folder_path, filename)
            print(f"File dengan nama yang mirip ditemukan di folder '{folder_name}': {file_path}")

            # Baca teks dari file DOCX dan proses
            text = read_text_from_docx(file_path)
            preprocessed_text = preprocess_text(text)
            tokens = tokenize_text(preprocessed_text)
            word_frequency = calculate_word_frequency(tokens)

            print("\n=== Kata-kata yang Paling Sering Muncul ===")
            print(word_frequency.most_common(10))

            # Menyimpan hasil frekuensi kata-kata pada berkas teks (txt)
            output_file_path = f'C:/Users/user/Untitled Folder/Destinasi/{folder_name}/frekuensi_kata_33226000{absen}_{folder_name}.txt'
            save_word_frequency_to_txt(word_frequency, output_file_path)

            print(f"Hasil frekuensi kata-kata setelah stemming dan preprocessing telah disimpan dalam berkas '{output_file_path}'.\n")
```
